```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('film_corpus.csv')
df2=pd.read_csv('FilmcorpusWithYear.csv')
```

```python
sLength = len(df['title'])
```

```python
df['year'] = pd.Series(np.random.randn(sLength), index=df.index)
```

```python
df.to_csv('FilmcorpusWithYear.csv')
```

```python
year=[]
for x in range(len(df['title'])):
    yearFound=0
    for y in range(len(df['title'][x])):
        if df['title'][x][y]=="(":
            temp=df['title'][x][y+1:y+5]
            year.append(df['title'][x][y+1:y+5]) 
            yearFound=1
            
        if (yearFound):
            df2['year'][x]=temp
            yearFound=0
            break

       
```

```python
df2.to_csv('FilmcorpusWithYear_updated.csv')
```
